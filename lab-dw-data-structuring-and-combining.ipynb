Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# Your code goes here
def clean_function (df):
    df_clean = df.dropna()
    df_clean = df_clean.drop_duplicates()
    df_clean.columns = df_clean.columns.str.lower()

    return df_clean
```

```python
import pandas as pd
df_1= pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df_2 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv")
df_3 = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv")
```

```python
cleaned_df1 = clean_function(df_1)
```

```python
cleaned_df1
```

```python
clean_df2 = clean_function(df_2)
```

```python
clean_df2
```

```python
clean_df3 = clean_function(df_3)
```

```python
clean_df3
```

```python
cleaned_df1.rename(columns ={"st": "state"},inplace = True)
```

```python
clean_df2.rename (columns ={"st": "state" },inplace= True)
```

```python
clean_df_total = pd.concat([cleaned_df1, clean_df2, clean_df3], axis=0, ignore_index = True)
clean_df_total
```
